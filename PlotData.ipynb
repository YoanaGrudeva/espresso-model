```python
# Return flow rate based on weight in vial
# input: value of w (array or single value) in grams
# output: flow rate in m/s
def flow_from_weight(w):
    rho = 1014420.68 # density of espresso g/m^3
    R = 0.0295 # m; radius of bedz
    A = np.pi*R**2 #surface area of bed cross-section
    # find volume of vial
    # volume = weight/ density
    vol = w/rho
    # time per vial is 2s.
    q = vol/(2*A)
    return q

```

```python
# EXTRACT WEIGHT FROM DATA
w = []
start = 0
for i in range(13):
    exp = data.iloc[:,start:start+5]
    exp["Weight"] = pd.to_numeric(exp["Weight"], downcast="float")
    weight = np.array([x for x in exp["Weight"] if x ==x]) # removing NaN entries
    weight = weight[:-1]
    w.append(weight)
    start += 5
    
#convert to array
w = np.array(w)
```

```python
# MEASURE AVERAGE FIRST DRIP
drips = np.zeros(len(w))
flows = []
flow_lens = []
for i in range(len(w)):
    exp = w[i]
    first_vial_l = np.squeeze(np.where(w[i]!= 0))
    first_vial = first_vial_l[0] #vial index of first vial with liquid
    t_drip = first_vial*2 # first vial starts at t=0s, second at t = 2s, third at t = 4s.
    drips[i] = t_drip
    non_zero_weights = exp[first_vial:] # np array of non-zero weights
    flows.append(flow_from_weight(non_zero_weights))
    flow_lens.append(len(non_zero_weights))
av_drip = np.average(drips)
print("first drip average", av_drip)
flows = np.array(flows)

```

```python
m = max(flow_lens)
reform_flows = np.zeros((13,m))
for i in range(len(w)):
    flow = flows[i]
    while m-len(flow) != 0:
            flow = np.append(flow,np.nan)
    reform_flows[i] = flow
```

```python
av_flow = np.nanmean(reform_flows,axis=0)
av_flow
```

```python
#TYPICAL PROFILE AFTER FIRST DRIP
plt.plot(av_flow)
plt.show()
```
